```python
def eh_duplicada(estado):
    dupla = estado.split(' ')
    if((len(dupla) > 1) and (dupla[0] == dupla[1])):
        return estado.split(' ')[0]
    else:
        return estado
    
df_variacao_cp.index = list(map(lambda x: eh_duplicada(x), df_variacao_cp.index))
df_variacao_cp
```
